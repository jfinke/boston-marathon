### Boston Marathon Scraping Engine
Object for scraping 2010-2016 Boston Marathon running data and storing in a MongoDB.

Methods:
- Query by lastname, and "next records"
- Can automatically "query all" or a subset
- Store raw data/responses in MongoDB, being careful to avoid duplicates
- Retrieve raw data/responses from MongoDB

Notes:
- Tested and found that this successfully scrapes 2010-2015 (March 2016).  
- Tested on 2016, works (April 19th, 2016)
- 2009 and earlier failed.
